```python
import pathlib
import os
```

```python
os.listdir(path_to_dataset)
```

```python
data = []
for img_path in os.listdir(path_to_dataset):
    country_id , year= img_path.split(".")[0].split("_")
    path = pathlib.Path(path_to_dataset , img_path).absolute()
    data.append([country_id , year , path])
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data ,columns=["country_id" , "year" , "path"])
```
